```python
import pandas as pd
kasus_fb = pd.read_csv('kasus-flu-burung-per-provinsi-2005-2012.csv', delimiter=",", encoding="utf-8")
```

```python
kasus_fb
```

# Deskripsi Data

Data diatas berisi mengenai kasus Flu Burung yang terjadi per provinsi. Data tersebut diambil pada tahun 2005-2012.
Data bersumber dari Ditjen PP & PL, Kementerian Kesehatan. 
Berikut ini merupakan penjelasan mengenai Variabel yang digunakan pada Dataset ini:

Penjelasan mengenai Variabel pada Dataset ini:
1.kode_provinsi: Kode Provinsi
2.nama_provinsi: Nama Provinsi
3.tahun: Tahun
4.jumlah_kasus: Jumlah kasus flu burung di provinsi tersebut
5.jumlah_meninggal: Jumlah yang meninggal karena flu burung di provinsi tersebut
6.cfr: Case Fatality Rate (persentase jumlah yang meninggal dibanding dengan jumlah kasus)

# Statistik Data

```python
# Jumlah penderita terbanyak pada kasus flu burung dalam satu provinsi
a = kasus_fb.jumlah_kasus.max()
print ("Jumlah penderita terbanyak dalam satu provinsi= ", a, "orang")

# Jumlah kematian terbanyak pada kasus flu burung dalam satu provinsi
b = kasus_fb.jumlah_meninggal.max()
print ("Jumlah meninggal terbanyak dalam satu provinsi= ", b, "orang")

# Presentase tertinggi Case Fatality Rate (CFR)
c = kasus_fb.cfr.max()
print ("Presentase tertinggi CFR= ", c, "%")
```

```python
# Jumlah kasus flu burung seluruh provinsi per tahun 2005-2012
kasus_fb.groupby('tahun').sum()
```

```python
# Rata-rata jumlah terjadinya kasus flu per provinsi selama tahun 2005-2012
kasus_fb.groupby('nama_provinsi').mean()
```

```python
# Rata-rata jumlah terjadinya kasus flu burung per tahun 2005-2012 
kasus_fb.groupby('tahun').mean()
```

# Visualisasi Data

```python
ax = kasus_fb.groupby('tahun').mean()['jumlah_kasus']
ax = ax.plot(kind='bar', title ="Rata-Rata Jumlah Penderita Flu Burung Berdasarkan Tahun", figsize=(8,8), legend=False, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = kasus_fb.groupby('nama_provinsi').mean()['jumlah_kasus']
ax = ax.plot(kind='bar', title ="Rata-Rata Jumlah Penderita Flu Burung Setiap Provinsi pada Tahun 2005-2012", figsize=(12,5), legend=False, fontsize=12)
ax.set_xlabel("Nama Provinsi", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

# Analisis

Dari hasil visualisasi nilai statistik pada data kasus Flu Burung yang terjadi pada tahun 2005-2012, 
dapat dilihat bahwa jumlah kasus Flu Burung pada tahun 2005 hingga 2012 mengalami penurunan secara drastis.
Pada tahun 2005-2009 terjadi kasus Flu Burung sebanyak 171, kemudian pada tahun selanjutnya yaitu tahun 2011 terjadi 12 kasus
dan pada tahun 2012 kasus Flu Burung mengalami penurunan menjadi 9 kasus.
Dari seluruh provinsi di Indonesia, daerah dengan jumlah kasus tertinggi terdapat pada Prov. DKI Jakarta yang pernah mencapai 17 kasus kisaran tahun 2005 hingga 2012.

Pada data ini dapat diambil kesimpulan bahwa kasus flu burung yang terjadi di beberapa provinsi di Indonesia cenderung mengalami penurunan tiap tahunnya. 
Hal ini menunjukkan bahwa tingkat kepedulian masyarakat terhadap kesehatan mulai meningkat, kebersihan lingkungan serta tata cara merawat unggas sudah menjadi perhatian masyarakat sehingga kasus penyakit seperti flu burung dapat dihindari.
